```python
def exportGraph(G,name):
    filename=name+'_linkList.tsv'
    file=open(filename,mode='w')
    file.write('source'+'\t'+'target'+'\n')
    for link in G.edges():
        file.write(str(link[0])+'\t'+str(link[1])+'\n')
    file.close()
```

```python
grandL = []
art = []
links = []
for i in x['tracks']['items']:
    l =[]
    artists = i['track']['artists']
    for i in artists:
        l.append(i['name'])
        if i['name'] not in art:
            art.append(i['name'])
    grandL.append(l)
for i in grandL:
#     print(list(combinations(i,2)))
    p = list(combinations(i,2))
    if len(p)>0:
        for element in p:
            links.append(element)
```
